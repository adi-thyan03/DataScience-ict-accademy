```python
# Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.get_dataset_names()
```

1 violinplot

```python


sns.violinplot(x='species', y='sepal_length', data=iris, hue='species')
plt.title('Violin Plot of Sepal Length by Species')
plt.show()

```

 2 swarmplot

```python
sns.swarmplot(x='species', y='sepal_length', data=iris, hue='species')
plt.show()
```

3 stripplot

```python
sns.stripplot(x='species', y='sepal_length', data=iris, hue='species')
plt.title('Strip Plot of Sepal Length by Species')
plt.show()
```

4  boxplot

```python
sns.boxplot(x='species', y='sepal_length', data=iris, hue='species')
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

5 barplot

```python
sns.barplot(x='species', y='sepal_length', data=iris, hue='species')
plt.title('Bar Plot of Speal Length by Species')
plt.ylabel('Mean Value')
plt.show()
```

6 countplot

```python
sns.countplot(x='species', data=iris, hue='species')
plt.title('Count Plot of Species')
plt.show()
```

7 heatmap


```python
coor_matrix = iris.drop('species',axis=1).corr()
sns.heatmap(coor_matrix, annot=True)
plt.title('Heatmap of Correlation Matrix of Iris Data')
plt.show()
```

8 scatter

```python
sns.scatterplot(x='sepal_length', y='sepal_width',hue= 'species', data=iris)
plt.title('Scatter Plot of Sepal Length vs Speal Width')
plt.show()
```

9  line chart

```python

sns.lineplot(x='species', y='sepal_length', data=iris, hue= 'species',linewidth='2')
plt.title('Line Chart of Species')
plt.show()
```
